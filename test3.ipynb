```python
# import dataframe

df = pd.read_csv("data/df_out1.csv")
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.head()
```

# Training model on all features

```python
X = df.iloc[:,1:]      
Y = df.iloc[:,:1]

#---perform a split---
random_state = 12
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=1/3)
```

```python
def get_best_params_for_Random_Forest_Regressor(train_x, train_y):
     param_grid_Random_forest_Tree = {
                "n_estimators": [10, 20, 30],
                "max_features": ["auto", "sqrt", "log2"],
                "min_samples_split": [2, 4, 8],
                "bootstrap": [True, False]
                                    }
     grid = GridSearchCV(RandomForestRegressor(), param_grid_Random_forest_Tree, verbose=3, cv=5)
     grid.fit(train_x, train_y)
     n_estimators = grid.best_params_['n_estimators']
     max_features = grid.best_params_['max_features']
     min_samples_split = grid.best_params_['min_samples_split']
     bootstrap = grid.best_params_['bootstrap']

     decisionTreeReg = RandomForestRegressor(n_estimators=n_estimators, max_features=max_features,
                                                         min_samples_split=min_samples_split,
                                                         bootstrap=bootstrap)

```

```python
def get_model(train_x, train_y, test_x, test_y):
    randomForestReg = get_best_params_for_Random_Forest_Regressor(train_x, train_y)
    prediction_randomForestReg = randomForestReg.predict(test_x)  # Predictions using the randomForestReg Model
    prediction_randomForestReg_error = r2_score(test_y, prediction_randomForestReg)
    return randomForestReg
```

```python
X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
rf = get_model(X_train, Y_train, X_test, Y_test)
```

```python
rf.fit(X_train, Y_train)
```

```python
Y_pred_train = rf.predict(X_train)
Y_pred_test = rf.predict(X_test)
```

```python
print(f'Train R2: {r2_score(Y_train, Y_pred_train)}')
```

```python
print(f'Test R2: {r2_score(Y_test, Y_pred_test)}')
```
